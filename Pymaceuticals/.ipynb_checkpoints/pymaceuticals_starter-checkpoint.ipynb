## Observations and Insights 


#### Three Observations from the Data
1. Generally, there are more male mice (50.4%) than female mice (49.6%)
2. Mice treated with Ketapril has the highest average tumor volume (~55.24)
3. For the Mice treated with Capomulin, the correlation coefficient of 0.8419363424694717 
    shows a strong positive relationship between mouse weight and average tumor volume 
    Therefore, an increase in their weight causes an increase in their average tumor volume.

```python
#  %matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
mouse_metadata
study_results
# Combine the data into a single dataset
combined_mousestudy_df = pd.merge(study_results, mouse_metadata, 
                                 how='left', on='Mouse ID')
# Display the data table for preview
combined_mousestudy_df.head()
```

```python
# Checking the number of mice, including the duplicate mouse ID.
number_miceCombined=combined_mousestudy_df['Mouse ID'].nunique()
number_miceCombined
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice=combined_mousestudy_df.loc[combined_mousestudy_df.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_mousestudy_df.loc[combined_mousestudy_df['Mouse ID']=='g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_mousestudy_df[combined_mousestudy_df['Mouse ID'].isin(duplicate_mice)==False]
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
number_miceClean=clean_data.nunique()['Mouse ID']
number_miceClean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean_stat = clean_data.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
mean_stat
median_stat = clean_data.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
median_stat
variance_stat = clean_data.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
variance_stat
std_stat = clean_data.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
std_stat
sem_stat = clean_data.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]
sem_stat


population_mean = clean_data['Tumor Volume (mm3)'].median()
print(population_mean)

# Assemble the resulting series into a single summary dataframe.
summary_stat_df = pd.DataFrame({
    "Mean": mean_stat,
    "Median": median_stat,
    "Variance": variance_stat,
    "Standard Deviation": std_stat,
    "SEM": sem_stat
})
summary_stat_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
clean_data.groupby("Drug Regimen").agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std','sem']})
```

## Quartiles, Outliers and Boxplots

```python
clean_data
```

```python
# Start by getting the last (greatest) timepoint for each mouse
regimen = clean_data.groupby("Mouse ID").last()["Drug Regimen"]
timepoint= clean_data.groupby("Mouse ID").last()["Timepoint"]
tumor_vol = clean_data.groupby("Mouse ID").last()["Tumor Volume (mm3)"]

# Put treatments into a list to get the tumor volume at the last timepoint 
# for the mice treated with the regimen on the list
treat_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
boxplot_df = pd.DataFrame({
    "Regimen": regimen,
    "Timepoint": timepoint,
    "Tumor Volume":tumor_vol
})
boxplot_df

# Get the tumor volume at the last timepoint for mice treated with 
# the four drug regimen in the treat_list
boxplot_df = boxplot_df.loc[boxplot_df["Regimen"].isin(treat_list)]
boxplot_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Use treatments list for for loop (and later for plot labels)
# treat_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data=[]

for i in treat_list:
    print(i)
    
# Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = boxplot_df.loc[boxplot_df['Regimen'] == i]
    tumor_volume = final_tumor_vol["Tumor Volume"]
    tumor_vol_data.append(tumor_volume)

# Calculate the IQR and quantitatively determine if there are any potential outliers
    quartiles = tumor_volume.quantile([.25, .5, .75])
    first_quartile = quartiles[0.25]
    second_quartile = quartiles[0.5]
    third_quartile = quartiles[0.75]

    iqr = third_quartile - first_quartile
    iqr

    lower_bound = first_quartile - (1.5*iqr)
    upper_bound = third_quartile + (1.5*iqr)

    print(lower_bound, upper_bound)

# Determine outliers using upper and lower bounds
    outliers = tumor_volume.loc[(tumor_volume > upper_bound) | (tumor_volume < lower_bound) ]
    outliers
    print(outliers)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Highlight the outliers by changing colour amd style
purple_diamond = dict(markerfacecolor='purple', marker='D')

bplot = plt.boxplot(tumor_vol_data,labels=treat_list, patch_artist=True,vert=True, flierprops=purple_diamond)

# Highlight the boxplot by changing colours
Four_colors=["red", "blue", "yellowgreen", "yellow"]
for patch, color in zip(bplot['boxes'], Four_colors):
        patch.set_facecolor(color)
# Set the textual properties of the boxplot
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume (mm3)")
plt.title("Boxplot of Final Tumor Volume of each Mouse across Four Drug Regimens")

# Show the grid lines
plt.grid(True)

# Show the boxplot 
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
columns = ["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]
capomulin_TW = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin", columns]
capomulin_TW

capomulin_avgTumor = capomulin_TW.groupby(["Mouse ID"]).mean()
weight = capomulin_avgTumor.iloc[:,1]
weight

capomulin_avgTumor
average_tumor= capomulin_avgTumor.iloc[:,0]
plt.scatter(weight, average_tumor)
plt.xticks(weight, rotation=90)
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.title("Capomulin: Average Tumor Volume vs. Mouse Weight")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight 
# and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(weight,average_tumor)
correlation
```

```python
# Perform a linear regression on weight versus 
# average tumor volume for the Capomulin Regimen
vc_slope, vc_int, vc_r, vc_p, vc_std_err = st.linregress(weight, average_tumor)
vc_r
```

```python
# Create equation of line to calculate 
# predicted average tumor volume for the Capomulin Regimen
vc_fit = vc_slope * weight + vc_int
vc_fit
equation = f"y={round(vc_slope, 2)}x + {round(vc_int,2)}"
equation
```

```python
# Plot the line graph on the scatter plot, 
# and show the equation of the line for the Capomulin Regimen
plt.scatter(weight, average_tumor)
plt.plot(weight,vc_fit,"--",color="red")
plt.xticks(weight, rotation=90)
plt.annotate(equation,(19,35), fontsize=12, color ="red")
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.title("Capomulin: Average Tumor Volume vs. Mouse Weight")
plt.show()
```
